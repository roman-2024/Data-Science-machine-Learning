### Parameters of `plt.scatter`

`plt.scatter` is a function in Matplotlib used to create a scatter plot. Below are the parameters of this function along with their descriptions:

| Parameter    | Description |
|--------------|-------------|
| `x`          | The data for the x-axis (list or array). |
| `y`          | The data for the y-axis (list or array). |
| `s`          | The size of the markers (scalar or array) (default: 20). |
| `c`          | The color of the markers (list or array). Color values can be names, hex codes, or numbers (default: 'b'). |
| `cmap`       | Colormap for mapping values to colors (e.g., 'viridis', 'plasma', 'inferno'). |
| `alpha`      | The transparency level of the markers (0.0 to 1.0) (default: 1.0). |
| `marker`     | The shape of the markers ('o', '^', 's', 'D', etc.) (default: 'o'). |
| `linewidths` | Width of the marker edges (default: None). |
| `edgecolors` | Color of the marker edges (default: 'face'). |
| `label`      | The label for the legend. |




```python
#Scatterplot and basic
plt.scatter(data["tip"],data["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(
    x=data["total_bill"], 
    y=data["tip"], 
    s=100, 
    c=data["size"], 
    cmap='viridis', 
    alpha=0.7, 
    marker='o', 
    edgecolors='w', 
    linewidths=0.5
)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Tip vs Total Bill')
plt.colorbar(label='Size')
plt.show()
```

```python
plt.figure(figsize=(10,6))
colors=data["tip"]
plt.scatter(data["total_bill"],data["tip"],
            c=colors,s=data["total_bill"],edgecolor="black",lw=1,alpha=0.85,cmap="Greens")
cbar=plt.colorbar()
cbar.set_label("Satisfaction")
plt.tight_layout()
plt.show()
```

### Parameters of `plt.pie`

`plt.pie` is a function in Matplotlib used to create pie charts. Below are the parameters of this function along with their descriptions:

| Parameter    | Description |
|--------------|-------------|
| `x`          | The data for the pie slices (list or array). |
| `labels`     | Labels for each slice (list or array) (default: None). |
| `colors`     | Colors for each slice (list or array of color names or codes) (default: None). |
| `autopct`    | Format string or function to label the percentage of each slice (default: None). |
| `startangle` | The angle by which the pie chart is rotated (degrees, default: None). |
| `explode`    | Fraction of the radius to offset each slice (list or array) (default: None). |
| `shadow`     | If True, draw a shadow beneath the pie (default: False). |
| `radius`     | The radius of the pie (default: 1). |
| `pctdistance`| Distance of the percentage labels from the center (default: 0.6). |
| `wedgeprops` | Dictionary of properties to set for the wedges (default: None). |


```python
sizes = data["day"].value_counts()
labels = data["day"].value_counts().index
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1,0.05, 0, 0)  # explode 1st slice

plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    autopct='%1.1f%%', 
    startangle=140, 
    explode=explode, 
    shadow=True, 
    radius=0.8
)
plt.title('Day of tips data')
plt.show()
```

## Box Plot Attributes and Parameters

### `matplotlib.pyplot.boxplot`

The `boxplot` function in Matplotlib is used to create box plots. Here are the main attributes, parameters, and functions:

#### Parameters:
- **x**: array-like or list of arrays
  - The input data. It can be a list or array of data to plot.
  
- **notch**: bool, default False
  - If `True`, will create a notched box plot.
  
- **vert**: bool, default True
  - If `True`, will create a vertical box plot. If `False`, will create a horizontal box plot.
  
- **patch_artist**: bool, default False
  - If `True`, will fill the box with color.

- **showmeans**: bool, default False
  - If `True`, will show the mean of the data on the box plot.

- **meanline**: bool, default False
  - If `True`, will show a line at the mean of the data.

- **showfliers**: bool, default True
  - If `True`, will show the outliers in the plot.

- **showbox**: bool, default True
  - If `True`, will show the box part of the plot.

- **whis**: float or string, default 1.5
  - Defines the range of the whiskers. It can be a float value or a string like 'range' to specify the whisker range.

- **bootstrap**: int, default None
  - Number of bootstrap iterations to use for the bootstrapped confidence intervals.

- **k_depth**: int, default None
  - The number of boxplot levels to draw (used when plotting data with nested boxes).

- **positions**: array-like, default None
  - The positions of the boxes on the x-axis.

- **widths**: float or array-like, default 0.5
  - The width of each box.

- **patch_artist**: bool, default False
  - If `True`, will fill the boxes with color.

#### Returns:
- **boxes**: list of `matplotlib.patches.PathPatch`
  - The boxes drawn in the plot.

- **whiskers**: list of `matplotlib.lines.Line2D`
  - The whiskers drawn in the plot.

- **caps**: list of `matplotlib.lines.Line2D`
  - The caps drawn at the ends of the whiskers.

- **medians**: list of `matplotlib.lines.Line2D`
  - The lines representing the median of each box.

- **fliers**: list of `matplotlib.lines.Line2D`
  - The points representing the outliers.

- **means**: list of `matplotlib.lines.Line2D`
  - The lines representing the mean of each box (if `showmeans=True`).


```python
plt.boxplot(data["total_bill"], notch=True, patch_artist=True, showmeans=True, vert=False)
plt.title('Box Plot Example')
plt.show()
```

## Area Plot Attributes and Parameters

### `matplotlib.pyplot.fill_between`

The `fill_between` function in Matplotlib is used to create area plots. Here are the main attributes, parameters, and functions:

#### Parameters:
- **x**: array-like
  - The x coordinates of the data points. It must have the same length as `y`.

- **y1**: array-like
  - The y coordinates of the lower boundary of the area.

- **y2**: array-like, optional
  - The y coordinates of the upper boundary of the area. If not provided, it defaults to zero.

- **where**: array-like or boolean, optional
  - A boolean array or condition that specifies where to fill. If not provided, fills between `y1` and `y2` for all x values.

- **interpolate**: bool, optional, default False
  - If `True`, interpolate where `where` is `True`.

- **color**: color or sequence of colors, optional
  - The color of the filled area. Can be a single color or a sequence of colors.

- **alpha**: float, optional
  - The transparency level of the filled area, between 0 (fully transparent) and 1 (fully opaque).

- **label**: str, optional
  - The label for the plot, used for the legend.

- **hatch**: str, optional
  - The pattern used for filling the area (e.g., '/', '\\', '|', '-', etc.).

- **edgecolor**: color, optional
  - The color of the edge line around the filled area.

- **linewidth**: float, optional
  - The width of the edge line around the filled area.

- **linestyle**: str, optional
  - The style of the line used for the boundary of the filled area (e.g., '-', '--', '-.', ':').

#### Returns:
- **PolyCollection**: 
  - A collection of polygons representing the filled area.


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5, label='Area between sin and cos')
plt.plot(x, y1, color='blue', label='sin(x)')
plt.plot(x, y2, color='red', label='cos(x)')
plt.title('Area Plot Example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

<h2>Matplotlib Heatmap Attributes, Parameters, and Functions</h2>

<p>A heatmap is a graphical representation of data where values are depicted by color. In Matplotlib, heatmaps can be created using the <code>imshow</code> or <code>pcolor</code> functions. Here are the key attributes, parameters, and functions used to customize heatmaps:</p>

<h3>1. <code>imshow</code> Function</h3>
<p>The <code>imshow</code> function is commonly used to display data as an image, i.e., on a 2D regular raster. It is often used to create heatmaps.</p>
<ul>
  <li><strong>arr</strong>: <code>array-like</code> - The data to be displayed. Should be a 2D array.</li>
  <li><strong>cmap</strong>: <code>str or Colormap</code> - Colormap instance or registered colormap name. Defines the color map to be used.</li>
  <li><strong>interpolation</strong>: <code>str</code> - Interpolation method to be used. Options include 'nearest', 'bilinear', 'bicubic', etc.</li>
  <li><strong>aspect</strong>: <code>str</code> - Aspect ratio of the plot. Options include 'auto', 'equal', and numerical values.</li>
  <li><strong>origin</strong>: <code>str</code> - Origin of the data. Options include 'upper' or 'lower'.</li>
  <li><strong>extent</strong>: <code>tuple</code> - Bounding box in data coordinates (left, right, bottom, top).</li>
  <li><strong>vmin</strong>: <code>float</code> - Minimum data value that corresponds to the colormap’s minimum value.</li>
  <li><strong>vmax</strong>: <code>float</code> - Maximum data value that corresponds to the colormap’s maximum value.</li>
  <li><strong>alpha</strong>: <code>float</code> - The alpha blending value, between 0 (transparent) and 1 (opaque).</li>
</ul>

<h3>2. <code>pcolor</code> Function</h3>
<p>The <code>pcolor</code> function creates a pseudocolor plot with a non-regular rectangular grid. It is another way to create heatmaps.</p>
<ul>
  <li><strong>C</strong>: <code>array-like</code> - The data to be displayed. Should be a 2D array.</li>
  <li><strong>cmap</strong>: <code>str or Colormap</code> - Colormap instance or registered colormap name.</li>
  <li><strong>norm</strong>: <code>Normalize</code> - Normalize object for scaling data values to the interval [0, 1].</li>
  <li><strong>vmin</strong>: <code>float</code> - Minimum data value that corresponds to the colormap’s minimum value.</li>
  <li><strong>vmax</strong>: <code>float</code> - Maximum data value that corresponds to the colormap’s maximum value.</li>
  <li><strong>edgecolors</strong>: <code>str</code> - Color of the edge lines of the cells. Options include 'none' or 'face'.</li>
  <li><strong>shading</strong>: <code>str</code> - Shading mode. Options include 'flat', 'nearest', or 'gouraud'.</li>
</ul>

<h3>3. Colorbar</h3>
<p>A colorbar can be added to a heatmap to provide a reference for the data values.</p>
<ul>
  <li><strong>matplotlib.colorbar.Colorbar</strong> - Use the <code>colorbar</code> function to add a colorbar to your plot. It takes the following parameters:</li>
  <ul>

    <li><strong>ax</strong>: <code>Axes</code> - The axes to which the colorbar will be added.</li>

    <li><strong>pad</strong>: <code>float</code> - Padding between the colorbar and the main plot.</li>
  </ul>
</ul>

<h2>Matplotlib Contour Plot Attributes, Parameters, and Functions</h2>

<p>A contour plot is a graphical representation of three-dimensional data in two dimensions. Contours are lines that represent constant values of a function. In Matplotlib, contour plots can be created using the <code>contour</code> and <code>contourf</code> functions. Here are the key attributes, parameters, and functions used to customize contour plots:</p>

<h3>1. <code>contour</code> Function</h3>
<p>The <code>contour</code> function creates a contour plot with contour lines. It is used to plot the level curves of a function.</p>
<ul>
  <li><strong>X</strong>: <code>array-like</code> - The x coordinates of the data points. Should be 2D array if <code>Y</code> is also 2D.</li>
  <li><strong>Y</strong>: <code>array-like</code> - The y coordinates of the data points. Should be 2D array if <code>X</code> is also 2D.</li>
  <li><strong>Z</strong>: <code>array-like</code> - The values at the data points.</li>
  <li><strong>levels</strong>: <code>array-like</code> - The contour levels to be drawn. If not provided, levels are automatically determined.</li>
  <li><strong>cmap</strong>: <code>str or Colormap</code> - Colormap instance or registered colormap name. Defines the color map to be used for filled contours.</li>
  <li><strong>colors</strong>: <code>str or list of str</code> - Colors to use for the contour lines.</li>
  <li><strong>linestyles</strong>: <code>str or list of str</code> - Line styles for the contour lines.</li>
  <li><strong>vmin</strong>: <code>float</code> - Minimum data value for color scaling.</li>
  <li><strong>vmax</strong>: <code>float</code> - Maximum data value for color scaling.</li>
  <li><strong>extend</strong>: <code>str</code> - Extend the contour lines beyond the range of the data. Options include 'neither', 'both', 'min', 'max'.</li>
</ul>

<h3>2. <code>contourf</code> Function</h3>
<p>The <code>contourf</code> function creates a filled contour plot. It is used to fill the areas between contour lines with colors.</p>
<ul>
  <li><strong>X</strong>: <code>array-like</code> - The x coordinates of the data points. Should be 2D array if <code>Y</code> is also 2D.</li>
  <li><strong>Y</strong>: <code>array-like</code> - The y coordinates of the data points. Should be 2D array if <code>X</code> is also 2D.</li>
  <li><strong>Z</strong>: <code>array-like</code> - The values at the data points.</li>
  <li><strong>levels</strong>: <code>array-like</code> - The contour levels to be drawn. If not provided, levels are automatically determined.</li>
  <li><strong>cmap</strong>: <code>str or Colormap</code> - Colormap instance or registered colormap name. Defines the color map to be used for filled contours.</li>
  <li><strong>vmin</strong>: <code>float</code> - Minimum data value for color scaling.</li>
  <li><strong>vmax</strong>: <code>float</code> - Maximum data value for color scaling.</li>
  <li><strong>extend</strong>: <code>str</code> - Extend the color range beyond the data limits. Options include 'neither', 'both', 'min', 'max'.</li>
</ul>

<h3>3. Contour Labels</h3>
<p>To add labels to contour lines, use the <code>clabel</code> function from the <code>matplotlib.contour</code> module.</p>
<ul>
  <li><strong>contour</strong>: <code>QuadContourSet</code> - The result of the <code>contour</code> function.</li>
  <li><strong>inline</strong>: <code>bool</code> - Whether to place labels inline or not.</li>
  <li><strong>fontsize</strong>: <code>int</code> - Font size for the labels.</li>
</ul>

<h3>

```python
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Contour plot
plt.figure()
contour = plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot')
plt.show()

# Filled contour plot
plt.figure()
contourf = plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(contourf)
plt.title('Filled Contour Plot')
plt.show()
```

<h2>Matplotlib Area Plot Attributes, Parameters, and Functions</h2>

<p>An area plot is a variation of a line plot where the area below the line is filled with color. It is useful for visualizing the magnitude of change over time. In Matplotlib, area plots can be created using the <code>fill_between</code> function. Here are the key attributes, parameters, and functions used to customize area plots:</p>

<h3>1. <code>fill_between</code> Function</h3>
<p>The <code>fill_between</code> function fills the area between two horizontal curves.</p>
<ul>
  <li><strong>x</strong>: <code>array-like</code> - The x coordinates of the data points.</li>
  <li><strong>y1</strong>: <code>array-like</code> - The y coordinates of the first curve.</li>
  <li><strong>y2</strong>: <code>array-like</code> - The y coordinates of the second curve. If not provided, the area between <code>y1</code> and 0 is filled.</li>
  <li><strong>where</strong>: <code>array-like or boolean</code> - A condition or boolean array to determine where the fill should occur.</li>
  <li><strong>color</strong>: <code>str or color</code> - The color of the filled area.</li>
  <li><strong>alpha</strong>: <code>float</code> - The alpha blending value, between 0 (transparent) and 1 (opaque).</li>
  <li><strong>interpolate</strong>: <code>bool</code> - Whether to interpolate the fill between the lines.</li>
  <li><strong>label</strong>: <code>str</code> - Label for the filled area plot.</li>
</ul>

<h3>2. Customizing Plots</h3>
<p>To customize the appearance of an area plot, you can use various Matplotlib functions:</p>
<ul>
  <li><strong>plt.plot</strong> - To plot lines on top of the area plot.</li>
  <li><strong>plt.title</strong> - To set the title of the plot.</li>
  <li><strong>plt.xlabel</strong> - To set the x-axis label.</li>
  <li><strong>plt.ylabel</strong> - To set the y-axis label.</li>

<h2>Matplotlib Hexbin Plot Attributes, Parameters, and Functions</h2>

<p>A hexbin plot is a type of 2D histogram that shows the count of data points falling into hexagonal bins. It is useful for visualizing the density of data points. In Matplotlib, hexbin plots can be created using the <code>hexbin</code> function. Here are the key attributes, parameters, and functions used to customize hexbin plots:</p>

<h3>1. <code>hexbin</code> Function</h3>
<p>The <code>hexbin</code> function creates a hexbin plot. It visualizes the count of data points in hexagonal bins.</p>
<ul>
  <li><strong>x</strong>: <code>array-like</code> - The x coordinates of the data points.</li>
  <li><strong>y</strong>: <code>array-like</code> - The y coordinates of the data points.</li>
  <li><strong>gridsize</strong>: <code>int</code> - The size of the hexagon grid. Larger values mean larger hexagons.</li>
  <li><strong>cmap</strong>: <code>str or Colormap</code> - Colormap instance or registered colormap name. Defines the color map for the hexagons.</li>
  <li><strong>reduce_C_function</strong>: <code>function</code> - Function to reduce the data to a single value for each bin. Default is <code>np.mean</code>.</li>
  <li><strong>mincnt</strong>: <code>int</code> - Minimum count to show a hexagon. Hexagons with counts below this value are not shown.</li>
  <li><strong>edgecolors</strong>: <code>str</code> - Color of the hexagon edges. Options include 'none' or 'black'.</li>
  <li><strong>linewidths</strong>: <code>float</code> - Width of the hexagon edges.</li>
  <li><strong>alpha</strong>: <code>float</code> - The alpha blending value, between 0 (transparent) and 1 (opaque).</li>
</ul>

<h3>2. Customizing Plots</h3>
<p>To customize the appearance of a hexbin plot, you can use various Matplotlib functions:</p>
<ul>
  <li><strong>plt.title</strong> - To set the title of the plot.</li>
  <li><strong>plt.xlabel</strong> - To set the x-axis label.</li>
  <li><strong>plt.ylabel</strong> - To set the y-axis label.</li>
  <li><strong>plt.colorbar</strong> - To add a colorbar to show the scale of the counts.</li>
</ul>

<h3>3. Example Code</h3>
<p>Here is an example of creating a hexbin plot using random data:</p>
<pre><code>


```python
x = np.random.randn(1000)
y = np.random.randn(1000)
# Hexbin plot
plt.figure()
hb = plt.hexbin(x, y, gridsize=30, cmap='Blues', mincnt=1, edgecolors='black')
plt.colorbar(hb, label='Count')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hexbin Plot Example')
plt.show()
```

<h2>Matplotlib Error Bar Plot Attributes, Parameters, and Functions</h2>

<p>An error bar plot is used to represent the uncertainty or variability in data. Error bars show the range of possible values for each data point. In Matplotlib, error bar plots can be created using the <code>errorbar</code> function. Here are the key attributes, parameters, and functions used to customize error bar plots:</p>

<h3>1. <code>errorbar</code> Function</h3>
<p>The <code>errorbar</code> function adds error bars to a plot. It can be used with line plots, scatter plots, and more.</p>
<ul>
  <li><strong>x</strong>: <code>array-like</code> - The x coordinates of the data points.</li>
  <li><strong>y</strong>: <code>array-like</code> - The y coordinates of the data points.</li>
  <li><strong>yerr</strong>: <code>array-like or float</code> - The vertical error bars. Can be a single value for constant error or an array for different error values.</li>
  <li><strong>xerr</strong>: <code>array-like or float</code> - The horizontal error bars. Can be a single value for constant error or an array for different error values.</li>
  <li><strong>fmt</strong>: <code>str</code> - The format string for the data points (e.g., 'o' for circles, '-' for lines).</li>
  <li><strong>ecolor</strong>: <code>str or color</code> - The color of the error bars.</li>
  <li><strong>elinewidth</strong>: <code>float</code> - The width of the error bars.</li>
  <li><strong>capsize</strong>: <code>float</code> - The size of the cap at the end of each error bar.</li>
  <li><strong>capthick</strong>: <code>float</code> - The thickness of the cap lines.</li>
  <li><strong>alpha</strong>: <code>float</code> - The alpha blending value, between 0 (transparent) and 1 (opaque).</li>
</ul>

<h3>2. Customizing Plots</h3>
<p>To customize the appearance of an error bar plot, you can use various Matplotlib functions:</p>
<ul>
  <li><strong>plt.plot</strong> - To plot lines or markers along with error bars.</li>
  <li><strong>plt.title</strong> - To set the title of the plot.</li>
  <li><strong>plt.xlabel</strong> - To set the x-axis label.</li>
  <li><strong>plt.ylabel</strong> - To set the y-axis label.</li>
  <li><strong>plt.legend</strong> - To add a legend to the plot.</li>
</ul>

<h3>3. Example Code</h3>


```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.2  # Constant error
# Error bar plot
plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='red', elinewidth=2, capsize=4, capthick=2, alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bar Plot Example')
plt.grid(True)
plt.show()
```
